```r
#9
movies_modificado <- movies %>%
pivot_longer(cols = c(Action, Animation,Comedy,Drama,Documentary,Romance,Short),
names_to = "genre",
values_to = "is_genre") %>%
filter(is_genre == 1)

head(movies_modificado)
```

```r
#9
# Relación que existe entre género y rating
graf1 <- ggplot(movies_modificado, aes(x = genre, y = rating)) +
geom_boxplot() +                  
labs(x = "Género", y = "Rating",title = "Ratings por género")

# El género más frecuente ordenado
movies_modificado2 <- movies_modificado %>%
group_by(genre) %>%
summarize(count = n()) %>%
arrange(desc(count))

# Promedio de budget
avg_budget_by_genre_year <- movies_modificado %>%
group_by(genre, year) %>%
summarize(avg_budget = mean(budget, na.rm = TRUE))

# Relación promedio de budget a lo largo de los años por género
graf5 <- ggplot(avg_budget_by_genre_year, aes(x = year, y = avg_budget, color = genre)) +
geom_line() + facet_wrap(~ genre) +
labs(x = "Año", y = "Promedio de Presupuesto", title = "Promedio de Presupuesto por Género a lo largo de los Años")

# Creación histograma
graf4 <- ggplot(movies_modificado2, aes(x = reorder(genre, -count), y = count)) +
geom_bar(stat = "identity", fill = "blue") +
labs(x = "Género", y = "Cuenta", title = "Histograma de Géneros (Ordenado)") 


# Relación que existe entre presupuesto y rating por género diferenciado por año
graf2 <-ggplot(movies_modificado, aes(x = budget, y = rating, color=year)) +
geom_point(position = position_jitter(w = 1, h = 0), size = 1) +                     
facet_wrap(~ genre) +             
labs(x = "Presupuesto", y = "Rating", title = "Rating vs Presupuesto por Género") 


# Relación presupuesto y rating diferenciados por género
graf3 <- ggplot(movies_modificado, aes(x = budget, y = rating, color=genre)) +
geom_point(position = position_jitter(w = 0.2, h = 0), size = 1) +
labs(x = "Presupuesto", y = "Rating", title = "Scatterplot de Budget vs. Rating") 


graf1
graf4
graf2
graf3
graf5
                
```

```r
## 10. Gráfica logarítmica
library(patchwork)
library(scales)

g1 <- ggplot(diamonds,aes(x=carat,y=price))+geom_point()

miGraf <- ggplot(diamonds,aes(x=carat,y=price)) + 
  geom_point(colour = 'cyan') + 
  scale_x_log10(breaks = trans_breaks("log10", function(x) 10^x),
                labels = trans_format("log10", math_format(10^.x))) + 
  scale_y_log10(breaks = trans_breaks("log10", function(x) 10^x),
                labels = trans_format("log10", math_format(10^.x))) +
  geom_smooth(method='lm') +
  annotation_logticks()

g1 + miGraf
```
