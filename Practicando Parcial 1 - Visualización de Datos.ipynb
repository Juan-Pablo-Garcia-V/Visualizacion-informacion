```r
graf <-ggplot() + coord_cartesian()+ scale_x_continuous() + scale_y_continuous() + layer(
    data=diamonds,
    mapping=aes(x=carat, y=price),
    stat="identity",
    geom="point",
    position=position_identity()) + geom_smooth(method="lm", se='false')
graf
```

```r
graf <- ggplot(data=diamonds, aes(x=carat, y=price)) + scale_x_continuous() + scale_y_continuous() + geom_point() + labs(x='poop', y='poop', caption='poop')
graf
```
