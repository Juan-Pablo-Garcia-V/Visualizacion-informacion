```r
install.packages('ggraph')
install.packages('igraph')
install.packages('ggplot2')
library(ggraph)
library(igraph)
library(ggplot2)
```

```r
# Crea lista de orillas a partir de la estructura jerárquica
d1 <- data.frame(from="Vicerrector", to=c("Ante Salcedo", "Fer Esponda",
"Francisco Perez"))
d2 <- data.frame(from=rep(d1$to, each=3), to=c("J.
Guadalupe","Mónica","Enrique", "Octavio","Eva","Pepe","Ceci","Luis","Renata"))
orillas <-rbind(d1,d2)
# Crea el objeto
g <- graph_from_data_frame(orillas)
#... y lo desplegamos
ggraph(g,'dendrogram', circular=TRUE)+
geom_edge_diagonal()+
geom_node_point()+
geom_node_text(aes( label=name)) +
theme_void()
```

```r
install.packages('treemapify')
library(treemapify)
```

```r
df <- read.csv('VentasHibridos.csv')
df$VENTAS <- gsub("[,\\s]", "", df$VENTAS, perl = TRUE)
df$VENTAS <- as.integer(df$VENTAS)
print(df)
head(df)
```

```r
ggplot(df,aes(area=VENTAS, fill=VENTAS,
label=paste(NOM_ABR, "\n", VENTAS),subgroup=REGION))+
geom_treemap()+
geom_treemap_subgroup_border()+
geom_treemap_subgroup_text(place="center",
grow=T,alpha=0.5,color="white",
fontface="italic", min.size=0)+
geom_treemap_text(color="white",place="topleft",
reflow=T) + scale_fill_distiller(palette = "Greens")
```
